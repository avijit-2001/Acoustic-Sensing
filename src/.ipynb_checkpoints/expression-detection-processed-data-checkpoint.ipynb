```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
```

```python
from keras.layers.core import Dense, Dropout, Activation, Flatten
from keras.layers.convolutional import Convolution2D, MaxPooling2D
```

```python
import os
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from keras.utils import np_utils
from PIL import Image
```

```python
path = os.getcwd()
dataset = path + '/processed_data'
```

Constants

```python
extension = {
    'anger': 'an',
    'disgust': 'di',
    'fear': 'fe',
    'happiness': 'ha',
    'neutral': 'ne',
    'sadness': 'sa',
    'surprise': 'su',
    'contempt': 'co'
}

inv_extension = {
    value: key for key, value in extension.items()
}

classes = [key for key in extension]

IMG_SIZE = 250
epochs = 20
batch_size = 1
NUM_CLASSES = len(classes)
```

Helping Functions

```python
def plot_sample(data, label):
    data = data / 255.0
    plt.imshow(data, interpolation="nearest")
    plt.xlabel(label)
    plt.show()
```

```python
def extract_image(image_path):
    basewidth = IMG_SIZE
    hsize = IMG_SIZE
    img = Image.open(image_path)
    img = img.resize((basewidth, hsize), Image.ANTIALIAS)
    data = np.asarray(img)
    return data
```

Getting all Data in X, y format

```python
os.chdir(dataset)

X, y = [], []
for image in os.listdir(dataset):
    
    expression = image[:2]
    if expression not in inv_extension:
        continue
        
    image_path = os.path.join(dataset, image)
    data = extract_image(image_path)
    X.append(data)
    y.append(classes.index(inv_extension[expression]))
    
X_input = np.array(X)/255.0
y_input = np_utils.to_categorical(y, NUM_CLASSES)

os.chdir(path)
```

Plotting the Sample Data

```python
# for index, data in enumerate(X):
#     plot_sample(data, classes[y[index]])
```

Train and Test Spliting

```python
n = len(X_input)
n_train = (int)(n * 0.8)
n_test = n - n_train

X_train, X_test = X_input[:n_train], X_input[n_train:]
y_train, y_test = y_input[:n_train], y_input[n_train:]
```

```python
print(len(X_train), len(X_test), n_train, n_test, n)
```

```python
input_shape = X_train[0].shape
model = models.Sequential()
```

```python
input_shape
```

Building The Convulation part

```python
model.add(Convolution2D(64,3,3, padding='same',input_shape=input_shape))
model.add(Activation('relu'))
model.add(Convolution2D(64, 3, 3))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
```

Building the Artificial Neural Network

```python
model.add(Flatten())
model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dense(NUM_CLASSES))
model.add(Activation('softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=["accuracy"])
```

```python
hist = model.fit(
    X_train, y_train, 
    epochs=10, 
    batch_size = batch_size
)
```

```python
model.save('EmotionRecognitionColaboratory_10.model')
```

```python
new_model = tf.keras.models.load_model('EmotionRecognitionColaboratory_10.model')
new_model.evaluate(X_test, y_test)
```
